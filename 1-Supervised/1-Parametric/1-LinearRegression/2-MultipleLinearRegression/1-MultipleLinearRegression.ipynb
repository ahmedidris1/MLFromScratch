```python
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split
```

```python
x_LSTAT = scale(df["LSTAT"])
x_RM = scale(df["RM"])
Price_scaled = scale(df["Price"])
```

```python
plt.scatter(Price_scaled, x_RM, label="RM")
plt.scatter(Price_scaled, x_LSTAT, label="LSTAT")
plt.xlabel('Average of rooms and low status population')
plt.ylabel('Price')
plt.legend()
```

```python
X = np.c_[ np.ones(x_RM.shape[0]), x_RM, x_LSTAT ]
Y = df['Price']
```

```python
class MultipleLinearRegression:
    def __init__(self, lr=.01, n_iter=1000):
        self.n_iter = n_iter
        self.lr = lr
    
    
    def fit(self, X, Y):
        n_samples, n_features = X.shape
        self.w = np.zeros(n_features)
        self.b = 0
        
        loss_history = []
        w_updates = []
        b_updates = []
        
        for _ in range(self.n_iter):
        
            y_hat = np.dot(X, self.w) + self.b

            loss = ( 1 / (2*n_samples) ) * np.dot( (Y - y_hat).T, (Y - y_hat) )
            loss_history.append(loss)

            dw = ( 1/n_samples ) * np.dot( X.T, (y_hat - Y) )
            db = ( 1/n_samples ) * np.sum(y_hat - Y)
            w_updates.append(dw)
            b_updates.append(db)

            self.w -= self.lr*dw
            self.b -= self.lr*db
            
        return loss_history, w_updates, b_updates
    
    def predict(self, X):
        return np.dot(X, self.w) + self.b
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2)
```

```python
multiLinRegModel = MultipleLinearRegression(lr=.001, n_iter=1500)
```

```python
loss_history, w_updates, b_updates = multiLinRegModel.fit(X_train, Y_train)
```

```python
def r2_measure(Y, Y_predicted):
    return 1 - ( np.dot((Y - Y_predicted).T, (Y - Y_predicted)) / np.dot( (Y - np.mean(Y)).T, (Y - np.mean(Y)) ) )
```

```python
r_2 = r2_measure(Y_test, y_predicted)
print(f'r_2: {r_2:.3f}')
```
